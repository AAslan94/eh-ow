```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
    "font.size" : 18,
    "lines.linewidth" : 2,
})
```

```python
data = np.load('snr_resultsA.npz')
snr_dB = data['default']
snr_opt = data['optimized']
position = data['position']
snr_diff = np.array([66.01075923, 54.7057067 , 46.56294919, 41.54869007, 37.57106442,
       31.8786643 , 30.8054158 , 29.01762264, 26.71411109, 28.8701027 ,
       28.88244284, 31.97825025, 33.7066357 , 37.71662281, 40.30575136,
       44.45880155, 47.09233026, 50.17614896, 53.1675483 , 56.02444377,
       58.0823313 ])
```

```python
plt.figure(figsize=(10,6))
plt.plot(position, snr_dB, 'o-', label='Default')
plt.plot(position, snr_opt, 's-', label='Optimized')
plt.plot(position, snr_diff, 'x-', label='Optimized - Diffuse')
plt.xlabel(r'$x$ [m]')
plt.ylabel(r'$SNR$ [dB]')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("snr_opt_sun.pdf", format="pdf", bbox_inches="tight", dpi=300)
plt.show()
```
