```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
    "font.size" : 22,
    "lines.linewidth" : 2,
})
```

```python
nsun_or = np.load("no_sun_to_use21.npz")
ns_noor = np.load("nosun_noor_21.npz")
```

```python
pos = nsun_or['positions']
theta = nsun_or['orientations'][:,0]
phi = nsun_or['orientations'][:,1]
area = nsun_or['orientations'][:,2]
```

```python
x = pos[:,0]
x
```

```python
theta_deg = np.degrees(theta)
phi_deg = np.degrees(phi)

# Create plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Left axis for theta
ax1.plot(x, theta_deg, 'b-o')
ax1.set_xlabel(r'$x$ [m]')
ax1.set_ylabel(r'$\theta$ [deg]', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Right axis for phi
ax2 = ax1.twinx()
ax2.plot(x, phi_deg, 'r-s')
ax2.set_ylabel(r'$\phi$ [deg]', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Optional: improve layout and add grid
ax1.grid(True, linestyle='--', alpha=0.5)
fig.tight_layout()
plt.savefig("angles_no_sun.pdf", format="pdf", bbox_inches="tight", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Plot optimized and default with distinct styles
plt.plot(x, area * 1e4, 
         label='Optimized Orientation', 
         color='tab:blue', 
         linewidth=2.5, 
         marker='o', 
         markersize=6)

plt.plot(x, areax, 
         label='Default Orientation', 
         color='tab:orange', 
         linewidth=2.5, 
         linestyle='--', 
         marker='s', 
         markersize=6)

# Labels and title
plt.xlabel(r'$x$ $[\mathrm{m}]$')
plt.ylabel(r'$A_\mathrm{PV}$ $[\mathrm{cm^2}]$')


# Grid, legend, and layout
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(fontsize=16, frameon=True, loc='best')
plt.tight_layout()

# Improve tick readability
plt.tick_params(axis='both', which='major', labelsize=14)
plt.savefig("opt_area_no_sun21.pdf", format="pdf", bbox_inches="tight", dpi=300)
plt.show()
```
