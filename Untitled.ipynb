```python
def sinartisi (x):
    a0 = 3.10974e-19
    a1 = -1.43654e-15
    a2 = 2.70105e-12
    a3 = -2.66571e-9
    a4 = 1.482177e-6
    a5 = -4.622185e-4
    a6 = 7.611403e-2
    a7 = -5.181545
    return a7 + a6 * x + a5 * x**2 + a4 * x**3 + a3 * x**4 + a2 * x**5 + a1 * x**6 + a0 * x**7
```

```python
sinartisi(1000)
```
